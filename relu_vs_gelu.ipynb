```python
# Plot the differences between the ReLU activation function and GELU activation function
import numpy as np
import matplotlib.pyplot as plt
plt.style.use(['../figures/style.mpl'])

def relu(x):
    return np.maximum(x, 0)

def gelu(x):
    return 0.5 * x * (1 + np.tanh(np.sqrt(2/np.pi) * (x + 0.044715 * x**3)))

x = np.linspace(-5, 5, 100)
y_relu = relu(x)
y_gelu = gelu(x)

# make figure square
fig, ax = plt.subplots()
fig.set_size_inches(7, 7)
plt.plot(x, y_relu, label='ReLU', linewidth=2, linestyle='--')
plt.plot(x, y_gelu, label='GELU', linewidth=2, linestyle='-')
plt.grid()
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
plt.xlim(-3, 3)
plt.ylim(-1, 3)
plt.xticks(np.arange(-3, 4, 1))
plt.yticks(np.arange(-1, 4, 1))
plt.legend()
# save to file
plt.savefig('../figures/gelu_vs_relu.pdf', bbox_inches='tight')
plt.show()
```
